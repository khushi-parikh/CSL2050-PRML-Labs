## a + b

```python
class Kmeans():
  def __init__(self, k, iter, centroids):
    self.k = k                  # number of clusters
    self.iter = iter            # max number of iterations
    self.centroids = centroids
    
  def dist(self,x1,x2):
    return (sum((x1 - x2)**2))**0.5

  def calc_centroids(self, clusters, X):
    new_centroids = []
    new_df = pd.concat([pd.DataFrame(X), pd.DataFrame(clusters, columns=['cluster'])],
                      axis=1)
    for c in set(new_df['cluster']):
        current_cluster = new_df[new_df['cluster'] == c][new_df.columns[:-1]]
        cluster_mean = current_cluster.mean(axis=0)
        new_centroids.append(cluster_mean)
    # print("NEW DF")
    # print(new_df)
    return new_centroids, new_df

  def find_closest_centroid(self, ic, X):
    curr = []
    for i in X:
      distance = []
      for j in ic:
        distance.append(self.dist(i, j))
      curr.append(np.argmin(distance))
    return curr

  def algo(self,X):
    for i in range(self.iter):
      centroids = np.array(self.centroids)
      print(centroids)
      get_centroids = self.find_closest_centroid(centroids,X)
      new_centroids, new_df = self.calc_centroids(get_centroids, X)
      # plt.figure()
      # plt.scatter(np.array(centroids)[:, 0], np.array(centroids)[:, 1], color='black')
      # plt.scatter(X[:, 0], X[:, 1], alpha=0.1)
      # plt.show()
      if np.all(np.array(self.centroids) - np.array(new_centroids) < 1e-5):
        break
      self.centroids = new_centroids
    return self.centroids, new_df
```

```python
no_images = []
for images in os.listdir("C:/Users/acer/Downloads/q3/no"):
    no_img = cv2.imread(r"C:/Users/acer/Downloads/q3/no/" + images, 0)
    no_imgr = cv2.resize(no_img, (100, 100))
    no_images.append(no_imgr.reshape(-1))

yes_images = []
for images in os.listdir("C:/Users/acer/Downloads/q3/yes"):
    yes_img = cv2.imread(r"C:/Users/acer/Downloads/q3/yes/" + images, 0)
    yes_imgr = cv2.resize(yes_img, (100, 100))
    yes_images.append(yes_imgr.reshape(-1))

no = np.array(no_images)
yes = np.array(yes_images)
q3_df = pd.DataFrame(np.concatenate((no, yes), axis=0))
q3_df.shape
```

```python
from sklearn.preprocessing import StandardScaler
stan_scaler = StandardScaler()
scaled_img = stan_scaler.fit_transform(q3_df)
scaled_img.std(axis=0)
```

```python
# print(pd.DataFrame(scaled_img).value_counts())
```

```python
X = scaled_img
Y = np.concatenate((np.zeros(no.shape[0]), np.ones(yes.shape[0])))
```

## b

```python
from sklearn.decomposition import PCA

pca = PCA()
pca_transformed = pca.fit_transform(X)

plt.scatter(pca_transformed[:, 0], pca_transformed[:, 1], alpha=0.8, c=Y)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

## c

```python
for j in range(2):
    fig, axs = plt.subplots(2, 5, figsize=(12, 5))
    # plt.suptitle(f"cluster: {j+1}")
    print("Cluster",j+1)
    for i in range(10):
        axs[i % 2][i % 5].matshow(q3_df.values[Y == j][i].reshape(100,100), cmap="gray")
    plt.show()
```

## d

```python
from sklearn.cluster import AgglomerativeClustering as AggClust
from sklearn.metrics import silhouette_score as ss

ac = AggClust()
ac.fit(X)

print("Predicted Class labels :")
print(np.unique(ac.labels_,return_counts=True))
print("Silhouette Score :", ss(X, ac.labels_))
```

```python
for j in range(2):
    fig, axs = plt.subplots(2, 5, figsize=(12, 5))
    # plt.suptitle(f"cluster: {j+1}")
    print("Cluster",j+1)
    for i in range(10):
        axs[i % 2][i % 5].matshow(q3_df.values[ac.labels_ == j][i].reshape(100,100),cmap="gray")
    plt.show()
```

## e

```python
from sklearn.cluster import KMeans 

km = KMeans(n_clusters=2, n_init=20, tol=0.00001)
km.fit(X)

print("Predicted Class labels :")
print(np.unique(km.labels_,return_counts=True))
print("Silhouette Score :", ss(X, km.labels_))
```

```python
for j in range(2):
    fig, axs = plt.subplots(2, 5, figsize=(12, 5))
    # plt.suptitle(f"cluster: {j+1}")
    print("Cluster",j+1)
    for i in range(10):
        axs[i % 2][i % 5].matshow(q3_df.values[km.labels_ == j][i].reshape(100,100),cmap="gray")
    plt.show()
```

```python
from sklearn.metrics import homogeneity_score as hs
from sklearn.metrics import completeness_score as cs
from sklearn.metrics import v_measure_score as vs

print(" \t\t Agglomerative \t\t Kmeans")
print("Silhouette : \t", ss(X, ac.labels_), "\t", ss(X, km.labels_))
print("Completeness : \t", cs(Y, ac.labels_), "\t", cs(Y, km.labels_))
print("Homogeneity : \t", hs(Y, ac.labels_), "\t", hs(Y, km.labels_))
print("V_measure : \t", vs(Y, ac.labels_), "\t", vs(Y, km.labels_))
```
