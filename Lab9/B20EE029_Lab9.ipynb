```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
q1_df = scaler.fit_transform(q1_df.iloc[:,1:])
q1_df = pd.DataFrame(q1_df)
q1_df
```

```python
x = q1_df.iloc[:, :]
x_num = q1_df.iloc[:, 1:].values
```

## a

```python
from sklearn.decomposition import PCA 
pca = PCA()
pca.fit_transform(x)
pca_variance = pca.explained_variance_

plt.figure(figsize=(8, 6))
plt.bar(range(13), pca_variance, alpha=0.5, align='center', label='individual variance')
plt.legend()
plt.ylabel('Variance ratio')
plt.xlabel('Principal components')
plt.show()
```

```python
pca2 = PCA(n_components=2)
pca2.fit(x)
x_red = pca2.transform(x)
x_red
```

```python
q1_df_red = pd.DataFrame(x_red)
```

```python
sns.set_style("whitegrid")
sns.pairplot(q1_df_red,size=3);
plt.show()
```

```python
## The clusters can be sepereated in 3 classes by observing the above graph.
```

## b

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x_red)
```

```python
print("Predicted Class labels :- ")
print(kmeans.labels_)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x_red[y_kmeans == 0, 0], x_red[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Class 0')
plt.scatter(x_red[y_kmeans == 1, 0], x_red[y_kmeans == 1, 1], s = 100, c = 'orange', label = 'Class 1')
plt.scatter(x_red[y_kmeans == 2, 0], x_red[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Class 2')

#Plotting the centroids of the cluster
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'black', label = 'Centroids')

plt.legend()
```

## c

```python
from sklearn.metrics import silhouette_score
```

```python
for i in range(2, 12):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit_predict(x_red)
    score = silhouette_score(x_red, kmeans.labels_, metric='euclidean')
    print("Silhouette Score for clusters =", i , ":", score)
```

```python
# The optimal value of k is 3 as it has the highest silhoutte score.
# A silhoutte score of 1 means perfect distinction of clusters.
```
